## Importando as bibliotecas

```python
import numpy as np
import matplotlib.pyplot as plt
```

O código tenta reproduzir o experimento de transmissão através de um link utilizando luz para transmitir com duração de 2 horas de transmissão a uma distância de 270 metros.
O parâmetro leva em consideração um feixe construído em Laguerre-Gaussian.

## Com coeficiente de atenuação de 0.25

```python

# Parâmetros do experimento
num_simulacoes = 10000      # Número de simulações Monte Carlo
duracao_transmissao = 7200 # Duração da transmissão em segundos (2 horas)
intensidade_inicial = 1.0  # Intensidade inicial da transmissão
coeficiente_atenuacao = 0.25  # Coeficiente de atenuação no espaço livre

# Função para simular uma única transmissão com decaimento
def simular_transmissao(distancia, intensidade_inicial, coeficiente_atenuacao):
    distancia_percorrida = 270
    intensidade = intensidade_inicial
    while distancia_percorrida < distancia:
        distancia_percorrida += np.random.exponential(1)  # Simulando um passo com distribuição exponencial
        atenuacao = np.exp(-coeficiente_atenuacao * distancia_percorrida)
        intensidade *= atenuacao
    return intensidade

# Função para calcular a média de concorrência
def calcular_media_concorrencia(intensidades):
    return np.mean(intensidades)

# Realizar as simulações Monte Carlo
resultados = []
for _ in range(num_simulacoes):
    intensidade_final = simular_transmissao(duracao_transmissao, intensidade_inicial, coeficiente_atenuacao)
    resultados.append(intensidade_final)

# Calcular a média de concorrência
media_concorrencia = calcular_media_concorrencia(resultados)

# Imprimir a média de concorrência
print("Média de Concorrência:", media_concorrencia)

# Plotar o histograma dos resultados
plt.hist(resultados, bins=30, density=True, alpha=0.6, color='g')
plt.xlabel('Intensidade Final')
plt.ylabel('Frequência Normalizada')
plt.title('Distribuição da Intensidade Final após ' + str(num_simulacoes) + ' Simulações')
plt.show()
```
