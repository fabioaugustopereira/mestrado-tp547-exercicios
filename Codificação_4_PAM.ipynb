##Importando as bibliotecas

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Definindo os Parâmetros

```python
# Parâmetros para a geração de amostras aleatórias de modulação por amplitude de pulso
# de 4 níveis, onde são escolhidos números entre 0 e 3 para cada amostra.

num_amostras = 100  # Número de amostras
amplitude = 1  # Amplitude do sinal
niveis = 4  # Número de níveis de amplitude

# Geração de amostras aleatórias de números inteiros uniformente distribuídos entre 0 e 3 ( 4 níveis )
amostras = np.random.randint(0, niveis, size=num_amostras)

# Mapeamento de amostras geradas para amplitude do sinal para cada pulso gerado
amplitudes = (2 * amostras - (niveis - 1)) * amplitude / (niveis - 1)

# Plotagem do sinal
plt.stem(range(num_amostras), amplitudes)
plt.xlabel('Amostras')
plt.ylabel('Amplitude')
plt.title('Sinal 4-PAM')
plt.show()
```
