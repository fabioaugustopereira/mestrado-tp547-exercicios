##Importando as bibliotecas

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Definindo os Parâmetros

```python
# Parâmetros para a geração de amostras aleatórias de modulação por amplitude de pulso
# de 4 níveis, onde são escolhidos números entre 0 e 3 para cada amostra.

num_amostras = 100  # Número de amostras
amplitude = 1  # Amplitude do sinal
niveis = 4  # Número de níveis de amplitude

# Geração de amostras aleatórias de números inteiros uniformente distribuídos entre 0 e 3 ( 4 níveis )
amostras = np.random.randint(0, niveis, size=num_amostras)

# Mapeamento de amostras geradas para amplitude do sinal para cada pulso gerado
amplitudes = (2 * amostras - (niveis - 1)) * amplitude / (niveis - 1)

# Plotagem do sinal
plt.figure(figsize=(15, 5))

plt.stem(range(num_amostras), amplitudes)
plt.xlabel('Amostras')
plt.ylabel('Amplitude')
plt.title('Sinal 4-PAM')
plt.show()
```

O sinal de modulação pela técnica 4-PAM (Pulse Amplitude Modulation com 4 níveis de amplitude) pode ser aplicado em vários exemplos de sistemas de comunicação digital:

1. **Redes de Comunicação Digital:** Em redes de comunicação digital,
como redes de computadores ou sistemas de telecomunicações, a modulação PAM é usada para representar dados digitais em sinais analógicos para transmissão. É uma variação comum usada para transmitir múltiplos bits por símbolo.
2. **Modems DSL (Digital Subscriber Line):** Em modems DSL, que fornecem conectividade de internet de alta velocidade através de linhas telefônicas, é usada para transmitir dados digitais sobre o par de cobre.
3. **Sistemas de Comunicação a Cabo:** Em sistemas de comunicação a cabo, como televisão a cabo e internet por cabo, pode ser usada em diferentes partes do sistema para transmitir dados digitais.
4. **Redes de Sensores sem Fio:** Em redes de sensores sem fio, onde os dispositivos precisam transmitir dados em um ambiente de baixa potência e de alcance limitado, pode ser usada para representar dados de forma eficiente.
5. **Comunicações de Satélite:** Em sistemas de comunicação via satélite, a modulação PAM pode ser usada em diferentes partes do sistema para transmitir sinais de voz, vídeo e dados.
6. **Comunicações de Áudio Digital:** Em sistemas de áudio digital, como CDs, DVDs e streaming de áudio, pode ser usada para representar amostras de áudio digital.
