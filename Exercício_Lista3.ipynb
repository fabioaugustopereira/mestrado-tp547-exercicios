```python

import numpy as np
import matplotlib.pyplot as plt
from numpy.linalg import matrix_power
n=np.arange(1,101)
x1=np.zeros((100))
x2=np.zeros((100))

for i in n:
		x0=[0.5,0.5]
		xt=[[0.5,0.5],[0.5,0.5]]
		xn=np.dot(x0,matrix_power(xt,i))


		x1[i-1]=xn[0]
		x2[i-1]=xn[1]

x1=np.append(x0[0],x1)
x2=np.append(x0[1],x2)
i=np.arange(0,101)
plt.figure(1)
plt.plot(i,x1)

plt.plot(i,x2)
plt.xlabel('passos')
plt.ylabel('probabilidade')
plt.grid()
plt.show()
```
