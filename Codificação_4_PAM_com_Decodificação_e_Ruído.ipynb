##Importando as bibliotecas

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Definindo os Par√¢metros

```python
# Par√¢metros para a gera√ß√£o de amostras aleat√≥rias de modula√ß√£o por amplitude de pulso
# de 4 n√≠veis, onde s√£o escolhidos n√∫meros entre 0 e 3 para cada amostra com ru√≠do gaussiano adicionado.

num_amostras = 100  # N√∫mero de amostras
amplitude = 1  # Amplitude do sinal
niveis = 4  # N√∫mero de n√≠veis de amplitude
ru√≠do_std = 0.1  # Desvio padr√£o do ru√≠do gaussiano

# Gera√ß√£o de amostras aleat√≥rias de n√∫meros inteiros uniformente distribu√≠dos entre 0 e 3 ( 4 n√≠veis )
amostras = np.random.randint(0, niveis, size=num_amostras)

# Mapeamento de amostras geradas para amplitude do sinal para cada pulso gerado
amplitudes_transmitidas = (2 * amostras - (niveis - 1)) * amplitude / (niveis - 1)

# Adi√ß√£o de ru√≠do gaussiano
ru√≠do = np.random.normal(0, ru√≠do_std, num_amostras)
amplitudes_recebidas = amplitudes_transmitidas + ru√≠do

# Detector de decis√£o
amostras_demoduladas = np.round((niveis - 1) * (amplitudes_recebidas / amplitude) / 2 + 0.5)

# Plotagem dos sinais transmitidos e recebidos
plt.figure(figsize=(15, 5))

plt.subplot(2, 1, 1)
plt.stem(range(num_amostras), amplitudes_transmitidas, linefmt='b-', markerfmt='bo', basefmt=' ')
plt.xlabel('Amostras')
plt.ylabel('Amplitude')
plt.title('Sinal Transmitido')
plt.grid(True)

plt.subplot(2, 1, 2)
plt.stem(range(num_amostras), amplitudes_recebidas, linefmt='r-', markerfmt='ro', basefmt=' ')
plt.xlabel('Amostras')
plt.ylabel('Amplitude')
plt.title('Sinal Recebido com Ru√≠do')
plt.grid(True)

plt.tight_layout()
plt.show()

# Compara√ß√£o entre amostras demoduladas e originais
erro = np.sum(amostras_demoduladas != amostras)
taxa_erro = erro / num_amostras
print(f'Taxa de erro de s√≠mbolo: {taxa_erro:.2%}')
```
